```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python

# STARTING VARIABLES
# orbits(years) -> revolutions(days) -> arcs(hours)
ORBITS_PER_SIMULATION   = 1 # years
REVOLUTIONS_PER_ORBIT   = 365 # days
ARCS_PER_REVOLUTION     = 24 # hours

# money to energy multiplier
# ie: for $1, you get 10 energy units
m2e = 50

# daily data points
X_revolutions           = list()
X_arcs                  = list()
money_earned            = list()
money_spent             = list()
e_consumed              = list()
e_expended              = list()
expended_vs_consumed    = list()
e_production            = list()
e_production_val       = list()
```

```python

class Me:
    energy  = 2000 # roughly equivalent to a traditional dietary calorie
    money   = 500

    def buy(self, damg):
        self.money -= damg
        return damage

    def consume(self, damg):
        e = (damg * m2e)
        self.energy += e
        return e

    def expend(self, hard_labor):
        hourly_expenditure = 80+hard_labor # "resting"-ish
        self.energy -= hourly_expenditure
        return hourly_expenditure

    def mine(self):
        # Mining logic / energy units mined per hr.
        # 169 lb of boneless lean meat for a 339 lb cow
        # @975 cal per lb => 1 kill = 164,775 cal
        # if it takes 24 hr. to hunt and kill:
        # ~6,866 cal produced per hour
        e_mined = np.random.randint(6000,7000)
        return e_mined

    def earn(self, amt):
        self.money += amt
        return amt
```

```python

# GAME LOOP

me = Me()

earnings = 0
expense = 0
consumption = 0
expenditure = 0
exp_v_cons = 0
production = 0
production_val = 0

for standard_orbit in range(ORBITS_PER_SIMULATION):

    for standard_revolution in range(REVOLUTIONS_PER_ORBIT):

        X_revolutions.append(int(str(standard_orbit)+str(standard_revolution)))

        for standard_arc in range(ARCS_PER_REVOLUTION):

            X_arcs.append(int(str(standard_orbit)+str(standard_revolution)+str(standard_arc)))

            stdarc_earning = 0
            stdarc_expense = 0
            stdarc_consumption = 0
            stdarc_expenditure = 0
            stdarc_production = 0

            # if getting hungry
            if me.energy <= 1200:
                # energy used to purchase and then eat
                # expend = me.expend(0)
                # stdarc_expenditure += expend
                stdarc_expenditure += me.expend(0)

                damage = 15 # basic value to pass to both functions
                # buy food
                # spend = me.buy(damage)
                # stdarc_expense += spend
                stdarc_expense += me.buy(damage)

                # eat the food
                # consume = me.consume(damage)
                # stdarc_consumption += consume
                stdarc_consumption += me.consume(damage)

            else:
                # or work and get money
                # expend = me.expend(95) # add 95 for hard labor
                # stdarc_expenditure += expend
                stdarc_expenditure += me.expend(95) # add 95 for hard labor

                # produce = me.mine() # mine more energy
                # stdarc_production += produce
                stdarc_production += me.mine() # mine more energy

                # earn = me.earn(25) # money earned per hour
                # stdarc_earning += earn
                stdarc_earning += me.earn(25) # money earned per hour

            earnings        += stdarc_earning
            expense         += stdarc_expense
            consumption     += stdarc_consumption
            expenditure     += stdarc_expenditure
            exp_v_cons      += abs(stdarc_expenditure - stdarc_consumption)
            production      += stdarc_production
            production_val  += (m2e * stdarc_production)

            money_earned.append(earnings)
            money_spent.append(expense)
            e_consumed.append(consumption)
            e_expended.append(expenditure)
            expended_vs_consumed.append(exp_v_cons)
            e_production.append(production)
            e_production_val.append(production_val)











```

```python

# Prepare variables for "per revolution" (per day) graphing
money_earned_per_rvl            = money_earned[::24]
money_spent_per_rvl             = money_spent[::24]
e_consumed_per_rvl              = e_consumed[::24]
e_expended_per_rvl              = e_expended[::24]
expended_vs_consumed_per_rvl    = expended_vs_consumed[::24]
e_production_per_rvl            = e_production[::24]
e_production_val_per_rvl        = e_production_val[::24]
```

```python
# PER REVOLUTION
# Income vs. Expenses
fig_a, ax_a = plt.subplots()
ax_a.set_title("Income vs. Expense / Rev")
ax_a.set_xlabel('Revolutions (aka: days)')
ax_a.set_ylabel('Income')

ax_a.plot(X_revolutions, money_earned_per_rvl, label='Income')
ax_a.plot(X_revolutions, money_spent_per_rvl, label='Expenses')

ax_a.legend()
fig_a.savefig('Graphs/income_vs_expense [revolutions].pdf')
# fig_a.show()
```

```python
# PER ARC
# Income vs. Expenses
fig_a, ax_a = plt.subplots()
ax_a.set_title("Income vs. Expense / Arc")
ax_a.set_xlabel('Arcs (aka: hours)')
ax_a.set_ylabel('Income')

ax_a.plot(X_arcs, money_earned, label='Income')
ax_a.plot(X_arcs, money_spent, label='Expenses')

ax_a.legend()
fig_a.savefig('Graphs/income_vs_expense [arcs].pdf')
# fig_a.show()
```

```python

# PER REVOLUTION
# ALL Direct Money Metrics
fig_b, ax_b = plt.subplots()
ax_b.set_title("Money Only / Rev")
ax_b.set_xlabel('Revolutions (aka: days)')
ax_b.set_ylabel('Money')

ax_b.plot(X_revolutions, money_earned_per_rvl, label='Money Made')
ax_b.plot(X_revolutions, money_spent_per_rvl, label='Money Spent')
ax_b.plot(X_revolutions, e_production_val_per_rvl, label='$Energy Produced')

ax_b.legend()
fig_b.savefig('Graphs/monetary_values [revolutions].pdf')
fig_b.show()
```

```python

# PER ARC
# ALL Direct Money Metrics
fig_b, ax_b = plt.subplots()
ax_b.set_title("Money Only / Arc")
ax_b.set_xlabel('Arcs (aka: hours)')
ax_b.set_ylabel('Money')

ax_b.plot(X_arcs, money_earned, label='Money Made')
ax_b.plot(X_arcs, money_spent, label='Money Spent')
ax_b.plot(X_arcs, e_production_val, label='$Energy Produced')

ax_b.legend()
fig_b.savefig('Graphs/monetary_values [arcs].pdf')
fig_b.show()
```

```python

# PER REVOLUTION
# ALL Direct Energy Metrics
fig_c, ax_c = plt.subplots()
ax_c.set_title("Energy Only / Revolution")
ax_c.set_xlabel('Revolutions (aka: days)')
ax_c.set_ylabel('Energy')

ax_c.plot(X_revolutions, e_expended_per_rvl, label='Energy Expended')
ax_c.plot(X_revolutions, e_consumed_per_rvl, label='Energy Consumption')
ax_c.plot(X_revolutions, expended_vs_consumed_per_rvl, label='Expended vs Consumed')

ax_c.legend()
fig_c.savefig('Graphs/energetic_values [revolutions].pdf')
fig_c.show()
```

```python

# PER ARC
# ALL Direct Energy Metrics
fig_c, ax_c = plt.subplots()
ax_c.set_title("Energy Only / Arc")
ax_c.set_xlabel('Arcs (aka: hours)')
ax_c.set_ylabel('Energy')

ax_c.plot(X_arcs, e_expended, label='Energy Expended')
ax_c.plot(X_arcs, e_consumed, label='Energy Consumption')
ax_c.plot(X_arcs, expended_vs_consumed, label='Expended vs Consumed')

ax_c.legend()
fig_c.savefig('Graphs/energetic_values [arcs].pdf')
fig_c.show()
```

```python

# PER REVOLUTION
# Expended vs. Consumed
fig_d, ax_d = plt.subplots()
ax_d.set_title("Expended vs. Consumed / Revolution")
ax_d.set_xlabel('Revolutions (aka: days)')
ax_d.set_ylabel('Energy')

ax_d.plot(X_revolutions, expended_vs_consumed_per_rvl, label='Expended vs Consumed')

ax_d.legend()
fig_d.savefig('Graphs/energetic_expenditure_vs_consumption [revolutions].pdf')
fig_d.show()
```

```python

# PER ARC
# Expended vs. Consumed
fig_d, ax_d = plt.subplots()
ax_d.set_title("Expended vs. Consumed / Arc")
ax_d.set_xlabel('Arcs (aka: hours)')
ax_d.set_ylabel('Energy')

ax_d.plot(X_arcs, expended_vs_consumed, label='Expended vs Consumed')

ax_d.legend()
fig_d.savefig('Graphs/energetic_expenditure_vs_consumption [arcs].pdf')
fig_d.show()
```

```python

# PER REVOLUTION
# GRAPH TEMPLATE W/ BASE-VALS
clip_value = 25

Xdays = X_revolutions[:clip_value]

econsmd = e_consumed_per_rvl[:clip_value]
eexpend = e_expended_per_rvl[:clip_value]
expVcon = expended_vs_consumed_per_rvl[:clip_value]
eprod   = e_production_per_rvl[:clip_value]
eprodV  = e_production_val_per_rvl[:clip_value]

econsmd = list(map(lambda x: m.ceil((x/1000)),     econsmd))
eexpend = list(map(lambda x: m.ceil((x/1000)),     eexpend))
expVcon = list(map(lambda x: m.ceil((x/100)),      expVcon))
eprod   = list(map(lambda x: m.ceil((x/100000)),   eprod))
eprodV  = list(map(lambda x: m.ceil((x/1000000)),  eprodV))

fig_, ax_ = plt.subplots()
ax_.set_title("Me vs World: Energy / Revolution")
ax_.set_xlabel('Revolutions (aka: days)')
ax_.set_ylabel('Energy')

ax_.plot(Xdays, econsmd,    label='Energy Consumption / 1,000')
ax_.plot(Xdays, eexpend,    label='Energy Expended/ 1,000')
ax_.plot(Xdays, expVcon,    label='Energy Consumption / 100')
ax_.plot(Xdays, eprod,      label='Energy Produced / 100,000')
ax_.plot(Xdays, eprodV,     label='$Energy Produced / 1M')

ax_.legend()
fig_.savefig("Graphs/me vs world: all metrics [revolutions].pdf")
fig_.show()
```

```python

# PER ARC
# GRAPH TEMPLATE W/ BASE-VALS
clip_value = 25

Xarcs = X_arcs[:clip_value]

econsmd = e_consumed[:clip_value]
eexpend = e_expended[:clip_value]
expVcon = expended_vs_consumed[:clip_value]
eprod   = e_production[:clip_value]
eprodV  = e_production_val[:clip_value]

econsmd = list(map(lambda x: m.ceil((x/1000)),     econsmd))
eexpend = list(map(lambda x: m.ceil((x/1000)),     eexpend))
expVcon = list(map(lambda x: m.ceil((x/100)),      expVcon))
eprod   = list(map(lambda x: m.ceil((x/100000)),   eprod))
eprodV  = list(map(lambda x: m.ceil((x/1000000)),  eprodV))

fig_, ax_ = plt.subplots()
ax_.set_title("Me vs World: Energy / Arcs")
ax_.set_xlabel('Arcs (aka: hours)')
ax_.set_ylabel('Energy')

ax_.plot(Xarcs, econsmd, label='Energy Consumption / 1,000')
ax_.plot(Xarcs, eexpend, label='Energy Expended/ 1,000')
ax_.plot(Xarcs, expVcon, label='Energy Consumption / 100')
ax_.plot(Xarcs, eprod, label='Energy Produced / 100,000')
ax_.plot(Xarcs, eprodV, label='$Energy Produced / 1M')

ax_.legend()
fig_.savefig("Graphs/me vs world: all metrics [arcs].pdf")
fig_.show()
```
